# Extraccion de informacion HTML
Usando la libreria **BeautifulSoup** podemos obtener un arbol de nodos para el documento html

```python
import bs4
import pandas as pd
soup = bs4.BeautifulSoup(response.text, 'html.parser')

sections = soup.select('.blog_w3ls')

url_selector = soup.findAll(attrs={"class" : "card-img-bottom"})

name_selector = soup.select("h5 a")
affiliation_selector = soup.select("h6")
bio_selector = soup.select("div.pt-3 p")


url = pd.Series([url['src'].replace('png', 'webp').replace('jpg', 'webp').replace('PNG', 'webp') for url in url_selector])
names = pd.Series([name.get_text() for name in name_selector])
affiliations = pd.Series([affiliation.get_text().replace('Affiliation:','').replace('  ','').replace('\r\n','') for affiliation in affiliation_selector])
bios = pd.Series([bio.get_text().replace('\r\n','') for bio in bio_selector])

executive = pd.DataFrame({'names':names[0:8],'affiliation':affiliations[0:8],'url':url[0:8],'bio':bios[0:8]})
researcher = pd.DataFrame({'names':names[8:],'affiliation':affiliations[8:],'url':url[8:],'bio':bios[8:]})

print(researcher)
Export = executive.to_json ('executive.json')
Export2 = researcher.to_json ('researcher.json')





```

# Extrayendo titulos de periodicos

```python
 import requests
import bs4

response2 = requests.get('https://www.eluniversal.com.mx/')
soup2 = bs4.BeautifulSoup(response2.text, 'html.parser')
```

```python
titlesH1 = soup2.select('h1.field-content a')
titlesH2 = soup2.select('h2.field-content a')

for title in titlesH1:
    print(title.get_atr())

titles = [title.get_text() for title in titlesH1] + [title.get_text() for title in titlesH2]

print(len(titles))

for title in titles:
    print(title)


```
