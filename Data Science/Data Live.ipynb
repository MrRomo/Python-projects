```python
import requests

response = requests.get('https://platzi.com/categorias/desarrollo/')

```

```python
print(response.text)
```

# Extraccion de informacion HTML
Usando la libreria **BeautifulSoup** podemos obtener un arbol de nodos para el documento html

```python
import bs4

soup = bs4.BeautifulSoup(response.text, 'html.parser')

courses_links = soup.select('.CarrersItem')

courses = [course['href'] for course in courses_links]

for course in courses:
    print(course)
```
